```r
#data.table
library(data.table)
library(tictoc)

tic("esercizio7")
#TASK1
tic("task1")
atac_peaks_dt <- fread("atac_peaks.bed.csv")

filtered_peaks <- atac_peaks_dt[
    chr == "chr2" & start >= 2000000 & start <= 4000000
]
toc()

#TASK2
tic("task2")
setorder(filtered_peaks, -score)
top_50_peaks <- head(filtered_peaks, 50)

toc()

print(top_50_peaks)

#fine timer generale
toc()
```

```r
# Data.frame
library(tictoc)

# Inizio timer generale
tic("esercizio7: data.frame")

# TASK 1: Caricamento e Filtro
tic("task1")

# 1. Caricamento del file CSV
# Usiamo read.csv, l'equivalente base di fread
atac_peaks_df <- read.csv("atac_peaks.bed.csv", stringsAsFactors = FALSE)

# 2. Filtro: Selezioniamo le righe che soddisfano le 3 condizioni
# Usiamo l'indicizzazione logica [rows, columns]
filtered_peaks_df <- atac_peaks_df[
    atac_peaks_df$chr == "chr2" & 
    atac_peaks_df$start >= 2000000 & 
    atac_peaks_df$start <= 4000000,
]
toc() # Chiusura timer Task 1

# TASK 2: Ordinamento e Selezione Top 50
tic("task2")

# 1. Ordinamento decrescente basato sulla colonna 'score'
# Usiamo order() per ottenere gli indici ordinati (il segno - è per l'ordine decrescente)
sorted_indices <- order(filtered_peaks_df$score, decreasing = TRUE)

# 2. Applicazione degli indici ordinati al dataframe e selezione delle prime 50 righe
top_50_peaks_df <- filtered_peaks_df[sorted_indices[1:50], ]

toc() # Chiusura timer Task 2

print("Risultato del Top 50 Peaks (data.frame):")
print(top_50_peaks_df)

# Fine timer generale
toc()
```

```r
#tabella confronto tempi
T_DT1 <- 0.076
T_DT2 <- 0.008
T_DF1 <- 0.032
T_DF2 <- 0.002

# Creazione della tabella riassuntiva
risultati_performance <- data.frame(
  Task = c("Task 1",
           "Task 2"),
  
  Tempo_data.table_Sec = c(T_DT1, T_DT2),
  Tempo_data.frame_Sec = c(T_DF1, T_DF2)
)

# Aggiungiamo una colonna per il fattore di velocizzazione (Speedup)
risultati_performance$Speedup_DT_vs_DF <- 
  round(risultati_performance$Tempo_data.frame_Sec / risultati_performance$Tempo_data.table_Sec, 1)

# Stampiamo la tabella finale
print(risultati_performance)
```
