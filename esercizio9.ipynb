```r
#data.frame
library(tictoc)
# install.packages("reshape2") # Esegui questo se non hai la libreria
library(reshape2) 

# Inizio timer generale
tic("esercizio9: data.frame")

# --- 1. Caricamento Dati ---
# Usiamo read.csv, l'equivalente base di fread
bulk_wide_df <- read.csv("bulk_counts_wide.csv", stringsAsFactors = FALSE)

# --- 2. Reshape (Melt) da Wide a Long ---
# dt_long <- melt(bulk_wide, id.vars = "gene", variable.name = "Sample", value.name = "Count")
df_long <- melt(bulk_wide_df,
                id.vars = "gene",
                variable.name = "Sample",
                value.name = "Count")

# --- 3. Calcolo Totale per Campione (Grouping) ---
# dt_long[, Sample_Total := sum(Count), by = Sample]
# Usiamo ave() per calcolare il totale del Conteggio (Count) per ogni Sample
df_long$Sample_Total <- ave(df_long$Count,
                            df_long$Sample,
                            FUN = sum)
print("Risultato Long Format con Sample_Total:")
print(head(df_long))


# --- 4. Calcolo della Media per Gene (Grouping e Aggregazione) ---
# dt_mean_wide <- dt_long[, .(Mean_Count = mean(Count)), by = gene]
df_mean_df <- aggregate(
    Count ~ gene,  # Formula: Aggrega 'Count' raggruppando per 'gene'
    data = df_long,
    FUN = mean
)
colnames(df_mean_df)[2] <- "Mean_Count" # Rinomina la colonna del risultato

# --- 5. Ordinamento ---
# setorder(dt_mean_wide, gene)
df_mean_df <- df_mean_df[order(df_mean_df$gene), ]


print("Risultato Mean Count per Gene:")
print(head(df_mean_df))

# Fine timer generale
toc()
```
