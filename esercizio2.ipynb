```r
#utilizzo data.frame
library(tictoc)
counts_df <- read.csv("bulk_counts_long.csv")
```

```r
#avvio timer generale
tic("esercizio2: data.frame")

tic("task1")
counts_df$log2_counts <- log2(counts_df$count + 1)
toc()
counts_df$high <- as.integer(counts_df$count > 100)
print("risulatato task 1")
print(head(counts_df))

tic("task2")
median_by_gene <- ave(counts_df$count, counts_df$gene, FUN = median)
counts_df$high <- as.integer(counts_df$count > median_by_gene)
toc()
print("Risultato TASK 2  con 'high' ricalcolato:")
print(head(counts_df))

#chiusura timer generale
toc()
```

```r
#tabella confronto tempi
T_DT1 <- 0.014
T_DT2 <- 0.056
T_DF1 <- 0.010
T_DF2 <- 0.018

# Creazione della tabella riassuntiva
risultati_performance <- data.frame(
  Task = c("Task 1",
           "Task 2"),
  
  Tempo_data.table_Sec = c(T_DT1, T_DT2),
  Tempo_data.frame_Sec = c(T_DF1, T_DF2)
)

# Aggiungiamo una colonna per il fattore di velocizzazione (Speedup)
risultati_performance$Speedup_DT_vs_DF <- 
  round(risultati_performance$Tempo_data.frame_Sec / risultati_performance$Tempo_data.table_Sec, 1)

# Stampiamo la tabella finale
print(risultati_performance)
```
